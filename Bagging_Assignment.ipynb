### Q1. How does bagging reduce overfitting in decision trees?

**Answer:**

Bagging (Bootstrap Aggregating) reduces overfitting by training multiple decision trees on different random subsets of the dataset (with replacement). Each model learns slightly different patterns, and their predictions are averaged (for regression) or voted (for classification). This reduces variance and prevents the model from fitting noise in the data, thus reducing overfitting.

### Q2. What are the advantages and disadvantages of using different types of base learners in bagging?

**Answer:**

Advantages:
- Flexible: Can use various models (trees, SVMs, etc.)
- Improved performance: Strong learners can give better accuracy
- Robustness: Reduces variance

Disadvantages:
- Complex models increase computation time
- Less interpretability
- Some learners may not benefit much from bagging (e.g., already low-variance models)

### Q3. How does the choice of base learner affect the bias-variance tradeoff in bagging?

**Answer:**

Bagging mainly reduces variance. If we use high-variance models (like decision trees), bagging significantly improves performance. Low-variance models (like linear regression) may not benefit much. The base learner determines whether bias or variance is dominant, and bagging primarily helps reduce variance without increasing bias much.

### Q4. Can bagging be used for both classification and regression tasks? How does it differ in each case?

**Answer:**

Yes, bagging can be used for both.

- Classification: Uses majority voting to decide the final class.
- Regression: Uses averaging of predictions.

The main difference is in how outputs are combined: voting vs averaging.

### Q5. What is the role of ensemble size in bagging? How many models should be included in the ensemble?

**Answer:**

Ensemble size refers to the number of models used in bagging.

- Larger size → better stability and reduced variance
- Too many models → increased computation without much gain

Typically, 50–200 models are used, depending on the dataset and resources. The performance improves until it stabilizes.

### Q6. Can you provide an example of a real-world application of bagging in machine learning?

**Answer:**

A common real-world example is Random Forest (a bagging-based method) used in:

- Fraud detection: Identifying fraudulent transactions
- Healthcare: Disease prediction
- Finance: Credit scoring

For example, in fraud detection, multiple decision trees are trained on different samples, and their combined output helps improve accuracy and reduce false positives.
